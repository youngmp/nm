# phase 1:1
The ODE below is in '=d/ds. 

```python
eps = .01
```

```python
def z1(th,order=3):
    zv=0;zh=0;zr=0;zw=0
    for i in range(order):
        zv += eps**(i+1)*system1.z['lam_'+'v0'][i](th)
        zh += eps**(i+1)*system1.z['lam_'+'h0'][i](th)
        zr += eps**(i+1)*system1.z['lam_'+'r0'][i](th)
        zw += eps**(i+1)*system1.z['lam_'+'qt0'][i](th)
        
    return np.array([zv,zh,zr,zw])
```

```python
def i1(th,order=3):
    iv=0;ih=0;ir=0;iw=0
    for i in range(order):
        iv += eps**(i+1)*system1.i['lam_'+'v0'][i](th)
        ih += eps**(i+1)*system1.i['lam_'+'h0'][i](th)
        ir += eps**(i+1)*system1.i['lam_'+'r0'][i](th)
        iw += eps**(i+1)*system1.i['lam_'+'qt0'][i](th)
        
    return np.array([iv,ih,ir,iw])
```

```python
def z2(th,order=3):
    zx=0;zy=0;zw=0
    for i in range(order):
        zx += eps**(i+1)*system2.z['lam_'+'x1'][i](th)
        zy += eps**(i+1)*system2.z['lam_'+'y1'][i](th)
        zw += eps**(i+1)*system2.z['lam_'+'qc1'][i](th)
        
    return np.array([zx,zy,zw])
```

```python
def i2(th,order=3):
    ix=0;iy=0;iw=0
    for i in range(order):
        ix += eps**(i+1)*system2.i['lam_'+'x1'][i](th)
        iy += eps**(i+1)*system2.i['lam_'+'y1'][i](th)
        iw += eps**(i+1)*system2.i['lam_'+'qc1'][i](th)
        
    return np.array([ix,iy,iw])
```

```python
def x1(th,order=3):
    v = system1.lc['lam_'+'v0'](th)
    h = system1.lc['lam_'+'h0'](th)
    r = system1.lc['lam_'+'r0'](th)
    w = system1.lc['lam_'+'qt0'](th)
    for i in range(order):
        v += eps**(i+1)*system1.g['lam_v0'][i](th)
        h += eps**(i+1)*system1.g['lam_h0'][i](th)
        r += eps**(i+1)*system1.g['lam_r0'][i](th)
        w += eps**(i+1)*system1.g['lam_qt0'][i](th)
    return np.array([v,h,r,w])
```

```python
def x2(th,order=3):
    x = system2.lc['lam_'+'x1'](th)
    y = system2.lc['lam_'+'y1'](th)
    w = system2.lc['lam_'+'qc1'](th)
    for i in range(order):
        x += eps**(i+1)*system2.g['lam_x1'][i](th)
        y += eps**(i+1)*system2.g['lam_y1'][i](th)
        w += eps**(i+1)*system2.g['lam_qc1'][i](th)
    return np.array([x,y,w])
```

```python
x2(10)
```

```python
# run without iso to start
def phase(t,y,system1,system2):
    t1,t2 = y
    x1_val = x1(t1)
    x2_val = x2(t2)
    vars_pair1 = [x1_val[0],x1_val[1],x1_val[2],x1_val[3],x2_val[0],x2_val[1],x2_val[2]]
    vars_pair2 = [x2_val[0],x2_val[1],x2_val[2],x1_val[0],x1_val[1],x1_val[2],x1_val[3]]
    dt1 = 1 + eps*np.dot(z1(t1),ct.coupling_thal(vars_pair1,system1.pardict,'val',system1.idx))
    dt2 = 1 + eps*np.dot(z2(t2),ct.coupling_cgl(vars_pair2,system2.pardict,'val',system2.idx))

    return np.array([dt1,dt2])
```

```python
dt = .01;T=10000
t = np.arange(0,T,dt)
```

```python
sol = solve_ivp(phase,[0,t[-1]],[0,0],method='LSODA',args=(system1,system2),t_eval=t)
```

```python
fig, axs = plt.subplots()
axs.plot(sol.t,sol.y.T[:,1]-sol.y.T[:,0])
```

# phase + isostable 1:1

```python
# run without iso to start
def phase_iso(t,y,system1,system2):
    t1,p1,t2,p2 = y
    x1_val = x1(t1)
    x2_val = x2(t2)
    vars_pair1 = [x1_val[0],x1_val[1],x1_val[2],x1_val[3],x2_val[0],x2_val[1],x2_val[2]]
    vars_pair2 = [x2_val[0],x2_val[1],x2_val[2],x1_val[0],x1_val[1],x1_val[2],x1_val[3]]
    dt1 = 1 + eps*np.dot(z1(t1),ct.coupling_thal(vars_pair1,system1.pardict))
    dp1 = 0
    dt2 = 1 + eps*np.dot(z2(t2),ct.coupling_cgl(vars_pair2,system2.pardict))
    dp2 = 0

    return np.array([dt1,dp1,dt2,dp2])
```

## compare to full model

```python
def ct_full(t,y):
    y1 = y[:4];y2 = y[4:]
    out1 = ct.rhs_thal(t,y1,system1.pardict,'val',0) + eps*ct.coupling_thal(y,system1.pardict,'val',0)
    out2 = ct.rhs_cgl(t,y2,system2.pardict,'val',1) + eps*ct.coupling_cgl(list(y2)+list(y1),system2.pardict,'val',1)
    return np.array(list(out1)+list(out2))
```

```python
sol_full = solve_ivp(ct_full,[0,t[-1]],[0,0,0,0,.333,0,0],method='LSODA',t_eval=t,)
```

```python
fig,axs = plt.subplots(figsize=(10,4))
axs.plot(sol_full.t,sol_full.y.T[:,0])
axs.plot(sol_full.t,sol_full.y.T[:,4])
#axs.set_xlim(sol_full.t[-1]-200,sol_full.t[-1])
axs.set_xlim(0,200)
```

```python
# estimate phase for each model and compare phase difference
skipn = 50
phase1 = np.zeros(len(sol_full.t[::skipn]))
phase2 = np.zeros(len(sol_full.t[::skipn]))

for i in range(len(sol_full.t[::skipn])):
    d1 = np.linalg.norm(sol_full.y.T[::skipn][i,:4]-system1.lc['dat'],axis=1)
    d2 = np.linalg.norm(sol_full.y.T[::skipn][i,4:]-system2.lc['dat'],axis=1)
    phase1[i] = np.argmin(d1)/len(system1.lc['dat'])
    phase2[i] = np.argmin(d2)/len(system2.lc['dat'])
```
